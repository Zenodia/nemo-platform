## Wrapped Dropdown Example

This cell is wrapped in a dropdown directive:

```python
# @nemo-nb: wrap dropdown Click to see advanced code
from typing import List, Optional


def process_data(data: List[int], filter_fn: Optional[callable] = None) -> List[int]:
    """Advanced data processing."""
    if filter_fn:
        return [x for x in data if filter_fn(x)]
    return data


result = process_data([1, 2, 3, 4, 5], lambda x: x % 2 == 0)
print(result)
```
