# Upload Evaluation Dataset to data store

In this notebook, we will upload the evaluation dataset to the data store.

## Prerequisites

- You need to have the `huggingface_hub` package installed.
- A running instance of the data store available at `http://data-store.test:8008`.

```python
from huggingface_hub import HfApi

hf_api = HfApi(endpoint="http://data-store.test:8008/v1/hf", token="token")
```

We use the `data` folder to upload all the test data. The folder contains the following files:
- simple-math.csv
- test_dataset_1.json

```python
dataset_name = "eval-test-data-math"
namespace = "default"
repo_id = f"{namespace}/{dataset_name}"

hf_api.create_repo(repo_id, repo_type="dataset")
```

```python
hf_api.upload_folder(folder_path="../datasets/math", path_in_repo="", repo_id=repo_id, repo_type="dataset")
```

Check the uploaded files:

```python
assert repo_id in [dataset.id for dataset in hf_api.list_datasets()]
```
