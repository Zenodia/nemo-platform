# Register entities in Entity Store

First, we will register the evaluation dataset in the entity store.

```python
import requests

res = requests.post(
    "http://nemo.test:8008/v1/datasets",
    json={"name": "math", "namespace": "default", "files_url": "hf://datasets/default/eval-test-data-math"},
)
dataset = res.json()
dataset
```

Next, we'll register the evaluation model.

```python
res = requests.post(
    "http://nemo.test:8008/v1/models",
    json={
        "name": "llama-3.1-8b",
        "namespace": "meta",
        "api_endpoint": {
            "url": "http://nim.test:8008/v1/completions",
            "model_id": "meta/llama-3.1-8b-instruct",
        },
    },
)
model = res.json()
model
```
